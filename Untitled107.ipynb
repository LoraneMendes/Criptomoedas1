#Funções necessárias

```python
def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        # find the end of this pattern
        end_ix = i + n_steps
        # check if we are beyond the sequence
        if end_ix > len(sequence)-1:
            break
        # gather input and output parts of the pattern
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return array(X), array(y)
```

###### Previsão Vanilla LSTM

```python
def classificar_valor(valor, percentis):
    if valor <= percentis[0]:
        return "Baixa Baixa"
    elif valor <= percentis[1]:
        return "Baixa Alta"
    elif valor <= percentis[2]:
        return "Média"
    elif valor <= percentis[3]:
        return "Alta Baixa"
    else:
        return "Alta Alta"

# Classificar os valores previstos usando a função classificar_valor
categorias_previstas_Vanilla_LSTM = [classificar_valor(valor, percentis) for valor in differences]

# Exibir as categorias previstas
print(categorias_previstas_Vanilla_LSTM)
```

```python

from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam

category_to_int = {"Baixa Baixa": 0, "Baixa Alta": 1, "Média": 2, "Alta Baixa": 3, "Alta Alta": 4}


X_train = np.array(differences).reshape(-1, 1)
y_train = [classificar_valor(valor, percentis) for valor in differences]
y_train_encoded = to_categorical([category_to_int[label] for label in y_train], num_classes=5)

model = Sequential()
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
optimizer.iterations


model.add(LSTM(50, activation='relu', input_shape=(n_steps, n_features)))
model.add(Dense(5))
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train_encoded, epochs=10, batch_size=32, validation_split=0.2)

predictions = model.predict(X_test)

# To get the most likely category for each prediction = np.argmax
predicted_classes = [np.argmax(pred) for pred in predictions]

# Convert the predicted class indices back to category labels
int_to_category = {v: k for k, v in category_to_int.items()}
predicted_categories = [int_to_category[index] for index in predicted_classes]
print(predicted_categories)
```
